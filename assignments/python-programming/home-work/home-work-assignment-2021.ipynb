## Question 3
<i>Write a fun which will take input as a dict and give me out as a list of all the values even in case of 2 level nesting it should work</i>

```python
# Sample input
people = {1: {'name': 'Tom', 'age': '32', 'sex': 'Male'},
          2: {'name': 'Marie', 'age': '28', 'sex': 'Female'}}
```

```python
# Function to convert dict to list
def dict_to_list(inp_dict, my_list):
    for value in inp_dict.values():
        if type(value) == dict:
            dict_to_list(value, my_list)
        else:
            my_list.append(value)
    return my_list
```

```python
# Test Case
dict_to_list(people, [])
```

## Question 5
<i>Write a function which will take multiple list as a input and give me concatnation of all the element as and output</i>

```python
# Function to join two list
def join_lists(*args):
    my_list = []
    for arg in args:
        my_list.append(arg)
    return ' '.join([str(val) for sublist in my_list for val in sublist])
```

```python
# Function to get list of files
def get_list_of_files(dir_path):
    # Check if directory exists
    if os.path.exists(dir_path):
        return [f for (root, dirs, file) in os.walk(dir_path) for f in file]
    else:
        return []
```

## Question 16
<i>Write a func to read a complete PDf file</i>

```python
# Function to read pdf file
def read_pdf(file_path):
    with open(file_path, 'rb') as pdf_file:
        pdf_reader = PdfFileReader(pdf_file)
        pages = pdf_reader.getNumPages()
        content = ''
        for index in range(pages):
            content += pdf_reader.getPage(index).extractText().replace('\n', '').strip() + '. '
        pdf_file.close()
        return content
```

```python
# Test Case
read_pdf('test_pdf.pdf')
```

## Question 20
<i>Write a function by which you will be able to append two PDF files</i>

```python
# Function to merge two PDF files
def merge_pdf_files(pdfs):
    pdf_merger = PdfFileMerger()
    for pdf in pdfs:
        pdf_merger.append(pdf)
    pdf_merger.write('merged_file.pdf')
    pdf_merger.close()
    return read_pdf('merged_file.pdf').strip()
```

```python
# Test Case
merge_pdf_files(['test_pdf.pdf', 'test_pdf_1.pdf'])
```
